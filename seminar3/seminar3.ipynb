Алгоритм шифрования называется  **симметричным**, если для шифрования и расшифровки используется один и тот же ключ. Если же ключи разные, то это **асимметричный**
В ассиметричном шифровании используются открытый и закрытый ключи. Открытый известен всем, а закрытый только владельцу.

Сценарии для асимметричного шифрования: 
  * Шифрование: кто угодно с открытым ключом может послать зашифрованное письмо владельцу ключа, но только владелец закрытого ключа может расшифровать это письмо
  * Подписание: владелец закрытого ключа может подписать сообщение, а кто угодно с открытым ключом может проверить подпись

Для симметричного шифрования сценарий подписи невалиден. Если знаешь ключ, которым проверяешь подпись, то и сам этим же ключом можешь подписать все что угодно и ключ теряет смысл, раз его знают все. А если не знаешь, то и не проверишь.

Так работает, если ключи известны. Но еще есть **сценарии взлома**, когда цель - расшифровать зашифрованное сообщение/ или подписать, не зная ключа. И смысл алгоритмов шифрования в том, чтобы сделать перебор всевозможных ключей заградительно дорогим по времени. Но пришли к этому не сразу (шифровать стали до появления вычислительных машин и тогда теоретическая сложность алгоритмов явно не оценивалась)

На семинаре рассмотрим несколько простейших симметричных алгоритмов шифрования

##### Шифр Цезаря
Есть алфавит `A` длины `M`, и сдвиг `offset` в качестве ключа.

**Шифрование**
Каждая буква `A[number]` в исходной строке c номером в алфавите `number` превращается в букву `A[(number + offset) % M]` в зашифрованной строке.

**Расшифровка**
Каждая буква `A[number]` в зашифрованной строке c номером `number` превращается в букву `A[(number - offset) % M]` в исходной строке.


Что с ним не так. Во-первых, вариантов ключа не так много. Во-вторых, буквы в алфавите по встречаемости в словах распределены неравномерно, и на основе частотного анализа можно восстановить ключ. (А. Конан-Дойл, "Пляшущие человечки")

Если сдвиг равен единице, то `python` -> `qzuipo`



##### Шифр Виженера
Есть алфавит `A` длины `M`, и ключ `key` - строка длины `N < M` .

**Шифрование**
Каждая буква `A[number]` в исходной строке c номером в алфавите `number` и позицией в строке `position` превращается в букву `A[(number + offset) % M]` в зашифрованной строке. Но `offset` вычисляется как сдвиг буквы `key[position % N]` относительно первой буквы алфавита.

**Расшифровка**

Каждая буква `A[number]` в зашифрованной строке c номером в алфавите `number` и позицией в строке `position` превращается в букву `A[(number - offset) % M]` в исходной строке. Но `offset` вычисляется как сдвиг буквы `key[position % N]` относительно первой буквы алфавита.

То есть, если у нас алфавит `abcdefghijklmnopqrtuvwxyz`, то ключ `a` - сохраняет исходную строку, произвольный однобуквенный ключ делает шифр Виженера шифром Цезаря. Ключ `ab` оставляет как есть буквы с четными номерами (начиная с 0), и сдвигает на единицу вправо буквы с нечетными номерами.

То есть `python` -> `pztioo`
